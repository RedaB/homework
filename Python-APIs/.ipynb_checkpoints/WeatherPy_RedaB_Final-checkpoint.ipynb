## WeatherPy_RedaB

## Dependencies & Setup

```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from citipy import citipy
from random import uniform
from config import api_key
```

## Generate Cities List

```python
# Generating the cities list:
cities = []

unique_cities = []
cities_name = []
countries_code = [] 

# Generating random coordinates (2500 cities)
for x in range(2500):
    lat, lng = np.random.uniform(low=-90.000, high=90.000), np.random.uniform(low=-180.000, high=180.000)
    city = citipy.nearest_city(lat, lng)
    cities.append(city)

# Generating unique cities
unique_cities = list(set(cities))
for x in range(800):
    unique_cities.append(unique_cities[x])
    cities_name.append(unique_cities[x].city_name)
    countries_code.append(unique_cities[x].country_code)
    
unique_cities_count = list(set(unique_cities))
print(len(unique_cities_count))
```

```python
# Creating a new data frame from the lists above
cities_df = pd.DataFrame({'City': cities_name,
                         'countries': countries_code})
cities_df.head()
```

## Perform API Calls

```python
# Looping through the list of cities_df then adding the data to a new df
url = "http://api.openweathermap.org/data/2.5/weather?"

column_names = ["City", "Country","Temp","Lat","Lon","Humidity","Clouds", "Wind Speed"]
data_df = pd.DataFrame(columns=column_names)

row_count = 1

print("Start Import")
for index, row in cities_df.iterrows():
    requested_city = row["City"]
    requested_url = url + "appid=" + api_key + "&units=IMPERIAL" + "&q=" + requested_city.replace(" ","+")
    response = requests.get(requested_url).json()
    
    if response["cod"] == "404":
        print("Data doesn't exist, please check next record!")

    else:
        data_df.at[index, "City"] = response["name"]
        data_df.at[index, "Country"] = response["sys"]["country"]
        data_df.at[index, "Temp"] = response["main"]["temp"]
        data_df.at[index, "Lat"] = response["coord"]["lat"]
        data_df.at[index, "Lon"] = response["coord"]["lon"]
        data_df.at[index, "Humidity"] = response["main"]["humidity"]
        data_df.at[index, "Clouds"] = response["clouds"]["all"]
        data_df.at[index, "Wind Speed"] = response["wind"]["speed"]       
        
        print("----------------------------------------------------------")
        print("Proceesing Record # " , row_count, ' | ' , response["name"])
        print(requested_url)
        row_count += 1
```

```python
# Count of generated data 
data_df.count()
```

## Exporting generated data into a CSV file

```python
# Exporting generated data into a csv file
data_df.set_index(['City'])
data_df.count()
```

```python
# Save data to a CSV file
data_df.to_csv("Resources/data.csv")
```

## Temperature (F) vs. Latitude

```python
# Creating plot
plt.scatter(data_df["Lat"], 
            data_df["Temp"], c=data_df["Temp"],
            edgecolor="blue", linewidths=1, marker="o", alpha=0.9, label="City")

plt.title("Temperature (F) vs. Latitude")
plt.ylabel("Max Temperature")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-70, 80])
plt.ylim([-10, 110])

# Save the figure
plt.savefig("Resources/Temperature_Latitude.png")

# Show plot
plt.show()
```

## Humidity (%) vs. Latitude

```python
# Creating plot
plt.scatter(data_df["Lat"], 
            data_df["Humidity"], c=data_df["Humidity"],
            edgecolor="orange", linewidths=1, marker="o", alpha=0.9, label="City")

plt.title("Humidity (%) vs. Latitude")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-70, 80])
plt.ylim([-10, 110])

# Save the figure
plt.savefig("Resources/Humidity_Latitude.png")

# Show plot
plt.show()
```

## Cloudiness (%) vs. Latitude

```python
# Creating plot
plt.scatter(data_df["Lat"], 
            data_df["Clouds"], c=data_df["Clouds"],
            edgecolor="green", linewidths=1, marker="o", alpha=0.9, label="City")

plt.title("Cloudiness (%) vs. Latitude")
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-70, 80])
plt.ylim([-10, 110])

# Save the figure
plt.savefig("Resources/Cloudiness_Latitude.png")

# Show plot
plt.show()
```

## Wind Speed (mph) vs. Latitude

```python
# Creating plot
plt.scatter(data_df["Lat"], 
            data_df["Wind Speed"], c=data_df["Wind Speed"],
            edgecolor="pink", linewidths=1, marker="o", alpha=0.9, label="City")

plt.title("Wind Speed (mph) vs. Latitude")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-70, 80])
plt.ylim([-10, 110])

# Save the figure
plt.savefig("Resources/WindSpeed_Latitude.png")

# Show plot
plt.show()
```

## Observations

1) It looks liek max temperature is the highest when Latitude is close to 0 (or near the equator). Cities that are located further out from the equator show lower max temperatures.


2) We can also apply the same obeservation above as far as humidity goes: the closer the cities are to the equator the highter is the humidity percentage.

3) As forr the cloudiness, it looks like it has the least correlation with Latitude compared with other variables. The slope is basically flat.
