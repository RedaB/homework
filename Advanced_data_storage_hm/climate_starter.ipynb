# 1- Reflecting Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
from sqlalchemy.sql import label, select
from flask import Flask, jsonify
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflecting an existing database into a new model
Base = automap_base()
# reflecting the tables
Base.prepare(engine, reflect=True)
```

```python
# Viewing all of the classes that automap found
Base.classes.keys()
```

```python
# Saving references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
inspector = inspect(engine)
columns = inspector.get_columns("measurement")
for c in columns:
    print(c["name"], c["type"])
    
```

```python
engine.execute("select * from measurement limit 10").fetchall()
```

```python
engine.execute("select * from station limit 10").fetchall()
```

# 2- Exploratory Climate Analysis

```python
# Retrieving the last 12 months of precipitation data and ploting the result
last_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()

# Calculating the date 1 year ago from today
last_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)

# Performing a query to retrieve the data and precipitation scores
year_ago = session.query(Measurement.date, Measurement.prcp).\
                        filter(Measurement.date > last_year).\
                        order_by(Measurement.date).all()
year_ago
```

```python
# Saving the results as DataFrame, setting the index to the date column and Sorting the dataframe by date
preciresult_df = pd.DataFrame(year_ago, columns=["Date", "Prcp"])

# preciresult_df.set_index("Date", inplace=True)
result_df = preciresult_df.dropna(how='any')
result_df.head()
```

```python
# Plotting with Matplotlib to plot the data
result_df.plot('Date', 'Prcp')
plt.grid(True)
plt.xlabel("Date")
plt.ylabel("Rain in Inches")
plt.title("Precipitation Analysis (8/24/16 to 8/23/17)")
plt.legend(["Precipitation"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# calcualting the summary statistics for the precipitation data
result_df.describe()
```

```python
# Defining the stations thatare available in this dataset
Station = session.query(Measurement.station).group_by(Measurement.station).count()
print("There are {} stations.".format(Station))
```

```python
# Defiing the most active stations
active_stations = session.query(Measurement.station, func.count(Measurement.tobs)).\
                     group_by(Measurement.station).\
                     order_by(func.count(Measurement.tobs).desc()).all()

# Defining the list of stations and the counts in descending order.
most_active_station = active_stations[0][0]
print("The most active station was",most_active_station,"with",active_stations[0][1],"observations.")
print()
print("Here are all of the sations (in descending order) with their observations:")
for station, count in active_stations:
    print("Station",station,"had",count, "observations.")
    
```

```python
# Defining the lowest/highest/average temperature recorded
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station == most_active_station).all()

```

```python
# Selecting the station with the highest number of temperature observations.
temperature = session.query(Measurement.station, Measurement.date, Measurement.tobs).\
    filter(Measurement.station == most_active_station).\
    filter(Measurement.date > last_year).\
    order_by(Measurement.date).all()

temp_df = pd.DataFrame(temperature)
temp_df.set_index("station", inplace=True)
temp_df.head()
```

```python
temp_df.plot.hist(by ='station', bins=12, color="pink")
plt.grid(True)
plt.title("Temperature Observations (8/24/16 to 8/23/17) for Station" + most_active_station, fontsize=10, fontweight="bold")
plt.xlabel("Recorded Temperature", fontsize=10, fontweight="bold")
plt.ylabel("Number of Recorded Observations", fontsize=10, fontweight="bold")
plt.legend(bbox_to_anchor=(1,1), fontsize=10)
plt.tight_layout()
#plt.savefig("Output/StationTemps.png")
plt.show()
```

```python
# Defining a function  `calc_temps` accepting start date and end date in the format '%Y-%m-%d' 
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
# Returning the minimum, average, and maximum temperatures for that range of dates 
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Calculating the tmin, tavg, and tmax for our trip
trip_arrive = dt.date(2018, 6, 1)
trip_leave = dt.date(2018, 6, 30)
last_year = dt.timedelta(days=365)
temp_avg_last_year = (calc_temps((trip_arrive-last_year), (trip_leave-last_year)))
print(temp_avg_last_year)
```

```python
# Plotting the results as a bar chart.
p2p = temp_avg_last_year[0][2]-temp_avg_last_year[0][0]
avg_temp = temp_avg_last_year[0][1]
min_temp = temp_avg_last_year[0][0]
max_temp = temp_avg_last_year[0][2]

fig, ax = plt.subplots()
bar_chart = ax.bar(1 , avg_temp, color= 'lightblue', yerr=p2p)
ax.set_xlabel("Trip")
ax.set_ylabel("Average Temperature")
ax.set_title("Trip Avg Temp")

def autolabels(rects):
    for rect in rects:
        h=rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., .6*h,'%.2f' % float(h) ,ha='center', va='bottom', fontsize=10)
        
# labelling the bars 
autolabels(bar_chart)        
plt.ylim(0, 100)
plt.xlim(0,2)
fig.tight_layout()
plt.show()
```

```python
# Calculating rainfall per weather station (sorting in descending order)
rainfall_by_station_lastyear = session.query(Measurement.station, func.sum(Measurement.prcp)).\
    filter(Measurement.date >= (trip_arrive-last_year)).\
    filter(Measurement.date <= (trip_leave-last_year)).\
    group_by(Measurement.station).\
    order_by(func.sum(Measurement.prcp).desc()).all()
rainfall_by_station_lastyear

```
